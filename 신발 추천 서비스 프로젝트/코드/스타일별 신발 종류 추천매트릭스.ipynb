```python
sum_formal=0
sum_casual=0
sum_feminine=0
for i in range(5):
    sum_formal+=x['formal'][i]
    sum_casual+=x['casual'][i]
    sum_feminine+=x['feminine'][i]
    
print(sum_formal)
```

```python
x = x.astype(float)
x['category']=x['category'].astype(int)
```
