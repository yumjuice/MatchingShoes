PIL -> byte Array

```python

import io

def PILtoArray(image_list):
    imgByteArrs=[]

    for image in image_list:
        imgByteArr = io.BytesIO()
        image.save(imgByteArr, format='PNG')
        
        imgByteArrs.append(imgByteArr.getvalue())
        print(imgByteArrs)
    return imgByteArrs
```

```python
import io
import matplotlib.pyplot as plt

classes=set()

imgByteArr = io.BytesIO()

for i,image in enumerate(image_list):
   
    if(image.format=='GIF'):
        continue
    imgByteArr = io.BytesIO()
    
    image.save(imgByteArr, format=image.format)
    print("image format",image.format)
    
    resp = req('/v2/vision/product/detect', 'POST',imgByteArr.getvalue())

    result=resp.json()
    print("result",result)
    width=result['result']['width']
    height=result['result']['height']
    checkBody=False
    checkShoes=False
    
    for object in result['result']['objects']:
        if object['class'] in ('one-piece','outer','shirts','t-shirts'):
            body=object['y1']*height
            checkBody=True
            print(body)
            continue
        elif object['class'] in ('shoes','sandals','sports shoes'):
            x1=object['x1']*width-50
            x2=object['x2']*width+50
            y1=object['y1']*height-30
            y2=object['y2']*height+30
            checkShoes=True
            continue
    if(checkBody and checkShoes):
        #신발사진 저장
        path='./img/formalShoes/'
        name=path+str(i)+'.jpg'
        img=image.crop((x1,y1,x2,y2))
        img.save(name)
        plt.imshow(img)
            
        #몸 사진 저장
        path2='./img/formalBody/'
        name2=path2+str(i)+'.jpg'
            
            
        bodyimg=image.crop((0,body,width,y1+40))
        bodyimg.save(name2)
        
        continue
        
    elif(checkBody):
        path2='./img/formalBodyOnly/'
        name2=path2+str(i)+'.jpg'
            
            
        bodyimg=image.crop((0,body,width,height))
        bodyimg.save(name2)

            
        continue
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import os
from PIL import Image

body_cascade=cv2.CascadeClassifier('./haarcascade_fullbody.xml')

def count_body(x):
    
    body=body_cascade.detectMultiScale(x,1.01,10,minSize=(30,30))
    print(body.shape[0])
    return body.shape[0]
```

```python
for img_name in os.listdir('./img/newcasual'):
    img=mpimg.imread('./img/newcasual/'+img_name)
    
    
    try:
        count_body(img)
        new_name='./img/newcasualFullBody/'+img_name
        mpimg.imsave(new_name,img)
        
    except:
        pass
    
```
