```python
import urllib.request

from urllib.parse import quote_plus #아스키 코드로 변환해준다

from bs4 import BeautifulSoup

from selenium import webdriver

import time

 

#해당 페이지를 크롤링 i가 페이지번호

def musinsaCrawling(pageNum):
    baseUrl = 'https://store.musinsa.com/app/items/lists/018/?category=&d_cat_cd=018&u_cat_cd=&brand=&sort=pop&sub_sort=&display_cnt=90'
    secondUrl='&page_kind=category&list_kind=small&free_dlv=&ex_soldout=&sale_goods=&exclusive_yn=&price=&color=11%2C9%2C47%2C12%2C53%2C58&a_cat_cd=&size=&tag=&popup=&brand_favorite_yn=&goods_favorite_yn=&blf_yn=&campaign_yn=&bwith_yn=&price1=&price2='
    pageUrl = '&page='

    url = baseUrl + pageUrl + str(pageNum) + secondUrl

    driver = webdriver.Chrome('C:/chromedriver')

    driver.get(url)

    time.sleep(3)  # 위에서 불러오고 1초 기다린후에 분석을 시작

 

    pageString = driver.page_source

    soup = BeautifulSoup(pageString, features="html.parser")


    
    
    #list=ul.find_all(name = 'li', attrs ={'class':'li_box'})
    ul = soup.find(name = 'ul', id='searchList')

    list = ul.find_all(name = "img")


    for i in list:

        try:
            image = i.attrs['data-original']

            reallink.append(image)

            #imgUrl = i.select_one('div.li_inner > div.list_img > a').img['src']
            print(image)

            #reallink.append(imgUrl)

        except:

            continue

 

    

    driver.close()

 

 

#plusUrl = input('검색할 옷을 입력하시오: ')

reallink = []

 
 

 
baseUrl = 'https://store.musinsa.com/app/items/lists/018/?category=&d_cat_cd=018&u_cat_cd=&brand=&sort=pop&sub_sort=&display_cnt=90'
secondUrl='&page_kind=category&list_kind=small&free_dlv=&ex_soldout=&sale_goods=&exclusive_yn=&price=&color=11%2C9%2C47%2C12%2C53%2C58&a_cat_cd=&size=&tag=&popup=&brand_favorite_yn=&goods_favorite_yn=&blf_yn=&campaign_yn=&bwith_yn=&price1=&price2='
        
pageUrl = '&page=1'

url = baseUrl + pageUrl + secondUrl

driver = webdriver.Chrome('C:/chromedriver')

driver.get(url)

 

time.sleep(3) #위에서 불러오고 3초 기다린후에 분석을 시작

 

 

pageString = driver.page_source

soup = BeautifulSoup(pageString, features="html.parser")

pageNum = int((soup.find("span",{"class" : "totalPagingNum"})).text)

print("pageNum", pageNum)

driver.close()

 

 

 

for i in range(1,pageNum+1):

    musinsaCrawling(i)

 


 

 


```
